```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd

def window_size():
    window = tk.Tk()
    window.wm_state('zoomed')
    window.config(background='sky blue')

    AMSA_Icon()
    DMC_Code_loading()
    inifile_loading()

    window.mainloop()

def AMSA_Icon():
    global file_label

    # create a label
    file_label = tk.Label(text=" ", background='sky blue')
    file_label.place(relx=0.2, rely=0.1)

    # create a button 
   
    button_font=("Calibri", 10)
    open_button = tk.Button(text='please input GD AMSA file\n请载入GD AMSA 文件', command=open_file, font=button_font)
    open_button.place(relx=0.05, rely=0.1)

    # create a button for edit
    
    edit_button = tk.Button(text='processing AMSA file\n处理AMSA文件', command=split_file, font=button_font)
    edit_button.place(relx=0.7, rely=0.1)


def DMC_Code_loading():
    global file_label_1

    # create a label
    file_label_1 = tk.Label(text=" ", background='sky blue')
    file_label_1.place(relx=0.2, rely=0.2)

    # create a button    
    button_font=("Calibri", 10)
    open_button = tk.Button(text='please input DMC code file\n请载入DMC code 文件', command=open_file1, font=button_font)
    open_button.place(relx=0.05, rely=0.2)

def inifile_loading():
    global inifile_lable
    #creat a lable 
    inifile_lable=tk.Label(text="",background='sky blue')
    inifile_lable.place(relx=0.2,rely=0.3)

    #create a button for inifile loading
    button_font=("Calibri", 10) 
    open_button=  tk.Button(text='please input ini. file\n请输入 ini 文件',command=open_file2,font=button_font)
    open_button.place(relx=0.05, rely=0.3)

    # create a button for edit
    
    edit_button = tk.Button(text='processing ini file\n处理ini文件', font=button_font)
    edit_button.place(relx=0.7, rely=0.3)
```

```python

def open_file():    

    global data

    file_path = filedialog.askopenfilename()
    file_label.config(text=file_path)
    data = pd.read_excel(file_path)
    return data

def open_file1():    

    global DMC_Code

    file_path = filedialog.askopenfilename()
    file_label_1.config(text=file_path)
    DMC_Code = pd.read_csv(file_path)
    return DMC_Code

def open_file2():    

    file_path = filedialog.askopenfilename()
    inifile_lable.config(text=file_path)
    inifile = pd.read_csv(file_path)
    return inifile

def split_file():

    global data
    global DMC_Code

    data[['id', 'DMC', 'temp1']] = data['chip_id'].str.split('=', expand=True)
    data[['DMC', 'Temp2']] = data['DMC'].str.split(',', expand=True)     


    DMC_Code=DMC_Code['DMC'].tolist()

    # data = data.loc[(data['DMC'] == '2237006AA00101008') | (data['DMC'] == '2237006AA00201004') | (data['DMC'] == '2237006AA00101005')]
    data = data.loc[data['DMC'].isin(DMC_Code)]

    data_columns = {
        '3001': '20Hz',
        '3002': '35Hz',
        '3003': '80Hz',
        '3004': '300Hz',
        '3005': '900Hz',
        '3006': '1000Hz',
        '3007': '1100Hz',
        '3008': '3000Hz',
        '3009': '8000Hz',
        '3010': '10000Hz',
        
    }
    with pd.ExcelWriter('AMSA.xlsx') as writer:
        for col, sheet in data_columns.items():
                data1 = data.loc[:,['site_num','DMC',f'{col};Output_{sheet}_94dBSPL_NPM;value;p']]
                data1.rename(columns={f'{col};Output_{sheet}_94dBSPL_NPM;value;p': 'SENS'}, inplace=True)

                data1.to_excel(writer, sheet_name=sheet, index=False)


# #按照DMC code filter

#     xls=pd.read_excel('AMSA.xlsx',sheet_name=None)C:\Users\wanglinl\Desktop\Linlin\2_Code_Temp\ACC_Project\AMSA.xlsx

#     for sheet_name, df in xls.items():
#         filtered_df= df[df['DMC'] == '2237006AA00200970']

#         with pd.ExcelWriter('AMSA.xlsx',mode='a') as writer:
#             filtered_df.to_excel(writer,sheet_name=sheet_name,index=False)


    tk.messagebox.showinfo("Success", "File processing completed!")


    

window_size()
```
